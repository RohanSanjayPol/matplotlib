```python
import seaborn as sns
```

```python
from matplotlib import pyplot as plt
```

```python
sns.set(style="whitegrid")
import pandas as pd
```

```python
p=pd.read_csv("Pokemon.csv")
```

```python
p.head()
```

```python
sns.barplot(x="Legendary",y="Speed",data=p)
plt.show()
```

```python
sns.barplot(x="Legendary",y="Speed",data=p,hue="Generation")
plt.show()
```

```python
sns.barplot(x="Legendary",y="Speed",data=p,palette='vlag')
plt.show()
```

```python
sns.barplot(x="Legendary",y="Speed",data=p,palette="rocket")
plt.show()
```

```python
sns.barplot(x="Legendary",y="Speed",data=p,color="green")
plt.show()
```

```python
a=pd.read_csv("player.csv")
```

```python
a.head()
```

```python
#scatterplot
sns.scatterplot(x="height_cm",y="weight_kg",data=a)
plt.show()
```

```python
sns.scatterplot(x="height_cm",y="weight_kg",data=a,hue="club",style="club")
plt.show()
```

```python
#distplot
sns.distplot(a["age"])
plt.show()
```

```python
sns.distplot(a["age"],hist=False)
plt.show()
```

```python
sns.distplot(a["age"],color="green")
plt.show()
```

```python
sns.distplot(a["age"],color="red",bins=10,kde=False)
plt.show()
```

```python
sns.distplot(a["age"],color="black",vertical=True)
plt.show()
```

```python
#jointplot
sns.jointplot(x="height_cm",y="weight_kg",data=a)
plt.show()
```

```python
sns.jointplot(x="height_cm",y="weight_kg",data=a,color="olive")
plt.show()
```

```python
sns.jointplot(x="height_cm",y="weight_kg",data=a,color="olive",kind="reg")
plt.show()
```

```python
#boxplot
sns.boxplot(x="club",y="weight_kg",data=a)
plt.show()
```
